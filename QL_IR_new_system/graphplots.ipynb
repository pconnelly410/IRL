```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import itertools as itools
import scipy as sc
from pathlib import Path
import random


HERE=Path.cwd()

```

```python
G = nx.Graph()

G.add_nodes_from([
     (1, {"color": "green"}),
     (2, {"color": "red"}),
     (3, {"color": "red"}),
     (4, {"color": "red"}),
 ])

G.add_edges_from([(1, 2), (1, 3), (2,3), (3,4)])
G.nodes[1]["color"] = "green"
```

```python
nx.draw(G, with_labels = True)
```

```python
posit = {1: np.array([0,0]), 2: np.array([1,1]), 3: np.array([1,-1]), 4: np.array([2,-1])}
nodes = nx.draw_networkx_nodes(G, posit, margins=0.5, node_color=['red', 'green', 'green', 'green'])
labels = nx.draw_networkx_labels(G, posit)
edges = nx.draw_networkx_edges(G, posit)
```

```python
positions = nx.spring_layout(G)
```
